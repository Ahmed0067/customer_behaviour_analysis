```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
 #create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python

# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)

```

```python

# Calculate the purchase count for each category
category_counts = df['category'].value_counts()

# Set up the visual dimensions
plt.figure(figsize=(9, 5))
sns.barplot(
    x=category_counts.index, 
    y=category_counts.values, 
    hue=category_counts.index, 
    palette="viridis", 
    legend=False
)

# Customize labels and titles
plt.title("Purchase Frequency Across Product Categories", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Number of Purchases", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)


# Render the graph
plt.tight_layout()
plt.show()
```

```python
frequency_matrix = pd.crosstab(df['category'], df['frequency_of_purchases'])
```

```python
plt.figure(figsize=(11, 6))
sns.heatmap(
    frequency_matrix, 
    annot=True,          
    fmt="d",             
    cmap="YlGnBu",       
    linewidths=0.8,      
    cbar=True            
)

# Label axes clearly using your column names
plt.title("Heatmap: Product Category vs. Frequency of Purchases", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Declared Purchase Frequency", fontsize=12)
plt.ylabel("Product Category", fontsize=12)

# Rotate labels slightly for perfect text readability
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
subscription_counts = df['subscription_status'].value_counts()
```

```python
# Set up the visual layout
plt.figure(figsize=(6, 6))

# Define clean, professional color palettes
colors = sns.color_palette('pastel')[0:len(subscription_counts)]
```

```python
# Generate the pie chart
plt.pie(
    subscription_counts.values, 
    labels=subscription_counts.index, 
    autopct='%1.1f%%',       # Shows the percentage with 1 decimal place
    startangle=140,          # Rotates the start for a cleaner look
    colors=colors, 
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5} # Adds clean white borders
)

# Set title details
plt.title("Proportion of Customer Subscription Status", fontsize=14, fontweight='bold', pad=20)

# Display the chart cleanly
plt.tight_layout()
plt.show()
```
